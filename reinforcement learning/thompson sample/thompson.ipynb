## Thompson sampling algorrithm

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv(r"C:\Users\sanid\OneDrive\Desktop\machine-learning-full\reinforcement learning\Ads_CTR_Optimisation.csv")
```

```python
import random
N = 10000
d = 10
ads_selected = []
no_rewards_1 = [0] * d
no_rewards_0 = [0] * d
total_reward = 0
for n in range(N):
    ad = 0
    max_thetha_i = 0
    for i in range(0,d):
        BETA = random.betavariate(no_rewards_1[i] + 1,no_rewards_0[i]+ 1)
        thetha_i = BETA
        if(thetha_i > max_thetha_i):
            max_thetha_i = thetha_i
            ad = i
    ads_selected.append(ad)
    reward = dataset.values[n,ad]
    if reward == 1:
        no_rewards_1[ad] = no_rewards_1[ad] + 1
    else:
        no_rewards_0[ad] = no_rewards_0[ad] + 1
total_reward = total_reward + reward
        

            
```

```python
plt.hist(ads_selected)
plt.title('Histogram of ads selections')
plt.xlabel('Ads')
plt.ylabel('Number of times each ad was selected')
plt.show()
```

Kmeans clusturing 

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

in clusturing we dont have label data so its unsupervised learning hence we need dot have independant and dependant variables


```python
data = pd.read_csv(r"C:\Users\sanid\OneDrive\Desktop\machine-learning-full\clusturing\Mall_Customers.csv")
```

for learning purposes we are using annaualincome and spending score to get a 2d plot to check the patterns in the data idealy we use all the features we need to predict the dependant variable

```python
X = data.iloc[:,[3,4]].values
```

ELBOW method (within clustur summ of squares to find ideal no of clusters)

init kmeans ++ to not fall into random intialization trap for the centroids
the first centroid is random placed but the next is weighted by the squared distance.

inertia_ = Sum of squared distances to centroids (lower = tighter clusters).

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,init="k-means++",random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title("THE ELBOW METHOD")
plt.xlabel("number of clusters")
plt.ylabel("WCSS")
plt.legend()
plt.show()
```

training the k means 

```python
kmeans = KMeans(n_clusters=5,init="k-means++",random_state=42)
y_kmeans = kmeans.fit_predict(X) #train and predict
```

```python
print(y_kmeans)
```

```python
plt.scatter(X[y_kmeans == 0 ,0],X[y_kmeans == 0 ,1],s=100,c='red',label ='Cluster 1') #cluster 1 plot 
plt.scatter(X[y_kmeans == 1 ,0],X[y_kmeans == 1 ,1],s=100,c='blue',label ='Cluster 2') #cluster 2 plot 
plt.scatter(X[y_kmeans == 2 ,0],X[y_kmeans == 2 ,1],s=100,c='green',label ='Cluster 3') #cluster 3 plot 
plt.scatter(X[y_kmeans == 3 ,0],X[y_kmeans == 3 ,1],s=100,c='cyan',label ='Cluster 4') #cluster 4 plot 
plt.scatter(X[y_kmeans == 4 ,0],X[y_kmeans == 4 ,1],s=100,c='magenta',label ='Cluster 5') #cluster 5 plot 

#plot the centroids to
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c='yellow',label = 'centroids') # 2d array
plt.title("clusters of customers")
plt.xlabel("annual income")
plt.ylabel("spending score (1-100)")
plt.legend()
plt.show()
```

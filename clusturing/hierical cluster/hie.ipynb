```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dt = pd.read_csv(r"C:\Users\sanid\OneDrive\Desktop\machine-learning-full\clusturing\hierical cluster\Mall_Customers.csv")
```

```python
X = dt.iloc[:,[3,4]].values
```

using dendograms to get optimal no of clustering

(a) sch.linkage(X, method='ward')
Purpose: Computes the linkage matrix, which defines how clusters are merged.

Input:

X → Your dataset (2D array, shape [n_samples, n_features]).

method='ward' → The linkage criterion (minimizes variance when merging clusters).

Common Linkage Methods
Method	Description	When to Use
'ward'	Minimizes variance (default)	Good for compact, spherical clusters
'single'	Minimum distance between clusters	Sensitive to noise
'complete'	Maximum distance between clusters	Prefers similar-sized clusters
'average'	Mean distance between clusters	Balanced approach

```python
import scipy.cluster.hierarchy as sch
dendogram = sch.dendrogram(sch.linkage(X,method='ward'))
plt.title('DENDOGRAM')
plt.xlabel('customers/observation points') # all the rows
plt.ylabel('Euclidien distances')
plt.legend()
plt.show()
```

check by drawing the horizontal bar and calculating the highest vertical line then the point of intersection at that thresshold is the no of clusters

```python
# Training the Hierarchical Clustering model on the dataset
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 5, metric='euclidean', linkage = 'ward')
y_hc = hc.fit_predict(X)

# Visualising the clusters
plt.scatter(X[y_hc == 0, 0], X[y_hc == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_hc == 2, 0], X[y_hc == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(X[y_hc == 3, 0], X[y_hc == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(X[y_hc == 4, 0], X[y_hc == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5')
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
dt = pd.read_csv(r"C:\Users\sanid\OneDrive\Desktop\machine-learning-full\regression models\random forest\Position_Salaries.csv")
```

```python
X = dt.iloc[:, 1:-1].values
Y = dt.iloc[:, -1].values
print(X)
print(Y)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=20,random_state=0)
model.fit(X,Y)
```

```python
model.predict([[6.5]])
```

```python
X_grid = np.arange(min(X), max(X), 0.1)           # From min to max with step 0.1
X_grid = X_grid.reshape((len(X_grid), 1))         # Convert to 2D for sklearn

# 3. Plotting
plt.figure(figsize=(10, 6))
plt.scatter(X, Y, color='red', label='Actual Data')  # Original data points
plt.plot(X_grid, model.predict(X_grid),         # Model predictions on grid
         color='blue', 
         label='Decision Tree Prediction')
plt.title('Salary vs Position Level (Decision Tree Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

tree_to_plot = model.estimators_[0]

# Plot the decision tree
plt.figure(figsize=(20, 10))
plot_tree(tree_to_plot, feature_names=dt.columns.tolist(), filled=True, rounded=True, fontsize=10)
plt.title("Decision Tree from Random Forest")
plt.show()
```

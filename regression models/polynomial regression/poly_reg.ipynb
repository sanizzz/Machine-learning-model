Polynomail linear regression

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv(r"C:\Users\sanid\OneDrive\Desktop\machine-learning-full\regression models\polynomial regression\Position_Salaries.csv")
print(dataset)
```

```python
X = dataset.iloc[:, 1:-1].values
Y = dataset.iloc[:, -1].values
print(Y)
```

TRAINING DATASET ON LINEAR REGRESSION AND PLOTTING


```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,Y)
```

TRAINIING POLY REGRESSIN MODEL ON WHOLE DATA SET with degree 4

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=7)
X_poly = poly_reg.fit_transform(X)
model_2 = LinearRegression()
model_2.fit(X_poly,Y)
```

visualizing linear regressin model

```python
plt.scatter(X,Y,color='red')
plt.plot(X,model.predict(X),color='blue')
plt.title('truthh or bluff {linear regression}')
plt.xlabel('position level')
plt.ylabel('salary')
plt.show()
```

polynomial regression visualizing results

```python
plt.scatter(X,Y,color='red')
plt.plot(X,model_2.predict(X_poly),color='blue')
plt.title('truth or bluff {poly linear reg}')
plt.xlabel('new position levels')
plt.ylabel('salaries')
plt.show()
```

```python
from sklearn.metrics import mean_absolute_percentage_error
mae_percent = mean_absolute_percentage_error(Y,model_2.predict(X_poly))
print(f"mae percentage score is {mae_percent}")
```
